# Source Splitting Tests

```python
import musdb
import numpy as np
import os
import ica_tools
import soundfile as sf
```

```python
mus = musdb.DB(root='inputs/musdb18')
num_tracks = 10
index = 0
for track in mus:
    if index >= num_tracks:
        break
    print(track.name)
    print(track.audio.shape)
    # get the mixture
    X = track.audio
    # get the sources
    sources = track.sources
    # get the source names
    source_names = [source for source in sources.keys()]
    # get the source audio signals
    s = np.array([sources[source].audio for source in source_names])
    # get the number of sources
    n_sources, n_samples, n_channels = s.shape[0], s.shape[1], s.shape[2]
    # reshape the sources to (n_sources, n_samples * n_channels)
    s = s.reshape((n_sources, n_samples * n_channels))
    # apply ICA
    #fast_ica(X: np.ndarray, out_channels: int, preprocess: str = "whiten", Cost: str = "Kurtosis", iterations: int = 256, threshold: float = 1e-6, PU: str = "CPU"):
    W, S = ica_tools.fast_ica(X, n_sources)
    print(f'W shape: {W.shape} X shape: {X.shape}')
    # apply the estimated mixing matrix to the sources
    S_est = X @ W.T
    # reshape the estimated sources to (n_sources, n_samples, n_channels)
    #S_est = S_est.reshape((n_sources, n_samples, n_channels))
    # save the estimated sources
    for i in range(n_sources):
        channel_audio = S_est[:, i]
        channel_audio = channel_audio / np.max(np.abs(channel_audio))
        channel_audio = channel_audio.astype(np.float32)
        sf.write(f'outputs/ica/{track.name}_{i}.wav', channel_audio, 44100)
    index += 1
```
